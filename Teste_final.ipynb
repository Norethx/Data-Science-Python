```python
import psycopg2
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical
import numpy as np
from tensorflow.keras.optimizers import Adam
```

```python
conexao = psycopg2.connect(host="localhost", dbname="credito", user="postgres", password="123456", port=5432)
```

```python
cursor = conexao.cursor()
#define consulta
consulta = 'SELECT "IDCREDITO","Duracao","HistoricoCredito","Proposito","Valor","Investimentos",COALESCE("Emprego",\'0\') "Emprego","TempoParcelamento","EstadoCivil","FiadorTerceiros",COALESCE("ResidenciaDesde",\'0\') "ResidenciaDesde","Idade","OutrosFinanciamentos",COALESCE("Habitacao",\'0\') "Habitacao","EmprestimoExistente","Profissao","Dependentes","SocioEmpresa","Estrangeiro","Status" FROM "CREDITO"'
cursor.execute(consulta)
```

```python
registros = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
Credito = pd.DataFrame(registros, columns = [desc[0] for desc in cursor.description])
```

```python
cursor.close()
conexao.close()
```

```python
Credito
```

```python
independente = Credito.iloc[:,1:19]
dependente = Credito.iloc[:,19:]

independente = np.array(independente)
dependente = np.array(dependente)
 
```

```python
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(independente,dependente, test_size = 0.3, random_state = 0)
```

```python
y_teste
```

```python
modelo = Sequential()
modelo.add(Dense(units = 9, input_dim = 18, activation='relu'))
modelo.add(Dense(units = 5,  activation='relu'))
modelo.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
modelo.summary()
```

```python
optimizer = Adam(learning_rate=0.001)
modelo.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])
modelo.fit(x_treinamento, y_treinamento, epochs = 1000, validation_data = (x_teste,y_teste))
```

```python
previsoes = modelo.predict(x_teste)
previsoes = (previsoes > 0.5).astype(int)
previsoes
```

```python
confusao = confusion_matrix(y_teste, previsoes)
```
