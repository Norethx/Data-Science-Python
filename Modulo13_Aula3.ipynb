```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Grafico de dispersão considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color='blue', facecolors = 'none', marker = '.')
plt.title('Árvores')
plt.xlabel('Circufêrencia')
plt.ylabel('Volume')
```

```python
plt.plot(base.Girth,base.Volume)
plt.title('Árvores')
plt.xlabel('Circuferencia')
plt.ylabel('Volume')
```

```python
#Grafico de dispersão com 'afastamento' dos dados jitter
#fit_reg linha de tendencia
sbn.regplot(x=base.Girth, y=base.Volume, data= base, x_jitter = 0.3, fit_reg = True)
```
