```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('co2.csv')
base.head()
base
```

```python
sbn.scatterplot(base, x = base.conc, y = base.uptake, hue = base.Type)
```

```python
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
plt.figure()
plt.subplot(1,2,1)
sbn.scatterplot(data = q, x = q.conc, y = q.uptake)
plt.title('Quebec')
plt.subplot(1,2,2)
sbn.scatterplot(data = m, x = m.conc, y = m.uptake)
plt.title('Mississippi')
plt.tight_layout()
```

```python
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
plt.figure()
plt.subplot(1,2,1)
sbn.scatterplot(ch, x = ch.conc, y = ch.uptake)
plt.title('chilled')
plt.subplot(1,2,2)
sbn.scatterplot(nc, x = nc.conc, y = nc.uptake)
plt.title('nonchilled')
plt.tight_layout()
```
