```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sbn
```

```python
#importação dos dados
df_dados = pd.read_csv('dados.csv', sep=';')
df_dados.head()
```

```python
#criar uma nova coluna com o esforço
df_dados['ESFORCO'] = df_dados['VALOREMPENHO'] / df_dados['PIB']
df_dados.head()
```

```python
#grafico de dispersão
sbn.scatterplot(data = df_dados, x='PIB', y='VALOREMPENHO')
plt.title('Cidades')
```

```python
#graficos de histograma e boxplot
plt.figure()
plt.subplot(2,2,1)
plt.boxplot(x = 'PIB', notch = True, vert=False, showfliers= False, data=df_dados, patch_artist=True)
plt.title('PIB')
plt.subplot(2,2,2)
plt.hist(df_dados['PIB'], bins= 10, color='green')
plt.subplot(2,2,3)
plt.boxplot(x = 'VALOREMPENHO', notch = True, vert=False, showfliers= False, data=df_dados, patch_artist=True)
plt.title('DADOSESFORCO')
plt.subplot(2,2,4)
plt.hist(df_dados['VALOREMPENHO'], bins= 10, color='blue')
plt.tight_layout()
```

```python
#criar um data frame com os 10 maiores valores de pib
MPIB = df_dados.nlargest(10,'PIB')
MPIB
```

```python
#Grafico de barras com os maiores PIBS
sbn.barplot(data=MPIB, x='MUNICIPIO', y='PIB', palette="inferno", hue_order=MPIB['MUNICIPIO'])
plt.xticks(rotation=45)
plt.title('Maiores PIBs')
```

```python
#criar um data frame com os 10 maiores valores de pib
MEMP = df_dados.nlargest(10,'VALOREMPENHO')
MEMP
```

```python
#Grafico de barras com os maiores PIBS
sbn.barplot(data=MEMP, x='MUNICIPIO', y='VALOREMPENHO', palette="viridis", hue_order=MEMP['MUNICIPIO'])
plt.xticks(rotation=45)
plt.title('Maiores empenhos')
```

```python
#criar um data frame com os 10 maiores valores de pib
MESF = df_dados.nlargest(10,'ESFORCO')
MESF
```

```python
sbn.barplot(data=MESF, x='MUNICIPIO', y='ESFORCO', palette="cubehelix", hue_order=MESF['MUNICIPIO'])
plt.xticks(rotation=45)
plt.title('Maiores PIBs por esforço gasto')
```
