```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df_dados = pd.read_csv('trees.csv')
df_dados.head()
```

```python
# Histograma com 10 divisões (bins) e somenta para o primeiro atributo da base de dados
plt.hist(df_dados.iloc[:,1], bins = 6)
```

```python
#Histograma com a linha de distribuição de frequencia, com 6 divisões (bins)
# kde = linha de densidade
sbn.histplot(df_dados.iloc[:,1], kde=False, bins = 6 , color='blue').set(title='Arvores')
```

```python
sbn.kdeplot(df_dados.iloc[:,1], color='blue').set(title='Arvores')
```

```python
sbn.histplot(df_dados.iloc[:,1], kde=True, bins=6, color='blue').set(title="Árvores")
```
